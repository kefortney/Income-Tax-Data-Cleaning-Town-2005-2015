```python
import pandas as pd
import numpy as np

# Visualisation
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns

# Configure visualisations
%matplotlib inline
```

```python
towntax = pd.read_csv("cumltowntaxincome.csv")
towntotal= pd.read_csv("cumltowntotal.csv")

```

```python
census = pd.read_csv("VT_2010_Census_Tract_Boundaries_and_Statistics.csv")
```

```python
towntax.head()
```

```python
towntotal.head()
```

```python
towntax = towntax.replace('*', np.NaN)
towntax = towntax.replace(' ', np.NaN)
```

```python
essex = towntotal[(towntotal['Town'] == 'Essex Junction') | (towntotal['Town'] == 'Essex Town')]
essex.head(25)
```

```python



numericlist = ['Return','Exempt','MarJnt','Single','MarSep','HdHous','AGI','FedTI','VTI','NetVTax']

for x in numericlist:
   essex[x] = pd.to_numeric(essex[x])
```

```python
essex.head(25)
```

```python
sns.barplot(x='Year' , y='Return' , hue='Town', data=essex , palette={'Essex Junction':'#0080ff', 'Essex Town':'#ffbf00'})
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Total Number of Returns')
plt.ylabel('Returns')
plt.show()
```

```python
sns.barplot(x='Year' , y='AGI' , hue='Town', data=essex , palette={'Essex Junction':'#0080ff', 'Essex Town':'#ffbf00'})
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ticklabel_format(style='plain', axis='y')
plt.title('Aggregated Adjusted Gross Income')
plt.ylabel('Returns')
```

```python
sns.barplot(x='Year' , y='NetVTax' , hue='Town', data=essex, palette={'Essex Junction':'#0080ff', 'Essex Town':'#ffbf00'})
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ticklabel_format(style='plain', axis='y')
plt.title('Net Vermont Taxes')
plt.ylabel('Returns')
```

```python
per = ['Exempt','MarJnt','Single','MarSep','HdHous']
essexperc = essex.copy()

for x in per:
    essexperc[x] = essexperc[x]/essexperc['Return']
```

```python
sns.pointplot(x='Year' , y='MarJnt' , hue='Town', data=essexperc, palette={'Essex Junction':'#0080ff', 'Essex Town':'#ffbf00'})
plt.title('Percentage of Married Joint Returns')
plt.ylabel('Percentage of Returns')
plt.ylim(0,1)
plt.show()

sns.pointplot(x='Year' , y='Single' , hue='Town', data=essexperc, palette={'Essex Junction':'#0080ff', 'Essex Town':'#ffbf00'})
plt.title('Percentage of Single Returns')
plt.ylabel('Percentage of Returns')
plt.ylim(0,1)
plt.show()
```
